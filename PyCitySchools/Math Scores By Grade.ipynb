```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#set school name as the index
school_df = school_data_complete.set_index("school_name")
school_df.head()
```

```python
ninth_df = school_df.loc[school_df["grade"] == "9th", :]

ninth_df = ninth_df.groupby("school_name")
ninth_df = ninth_df["math_score"].mean()

ninth_df
```

```python
tenth_df = school_df.loc[school_df["grade"] == "10th", :]
tenth_df = tenth_df.groupby("school_name")
tenth_math = tenth_df["math_score"].mean()
tenth_math
```

```python
eleventh_df = school_df.loc[school_df["grade"] == "11th", :]
eleventh_df = eleventh_df.groupby("school_name")
eleventh_math = eleventh_df["math_score"].mean()
eleventh_math
```

```python
twelvth_df = school_df.loc[school_df["grade"] == "12th", :]
twelvth_df = twelvth_df.groupby("school_name")
twelvth_math = twelvth_df["math_score"].mean()
twelvth_math
```

```python
Grade_df = pd.DataFrame([ninth_math, tenth_math, eleventh_math, twelvth_math])
Grade_df
```

```python
Grade_df = Grade_df.T
Grade = Grade_df.rename(columns = {"math_score":"9th", "math_score":"10th","math_score":"11th","math_score":"12th"})
Grade
```
