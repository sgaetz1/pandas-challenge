```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#set school name as the index
school_df = school_data_complete.set_index("school_name")
school_df.head()
```

```python
#find the average of 9th grade reading by school
ninth_df = school_df.loc[school_df["grade"] == "9th", :]
ninth_df = ninth_df.rename(columns = {"reading_score": "9th"})
ninth_df = ninth_df.groupby("school_name")
ninth_read = ninth_df["9th"].mean()
ninth_read
```

```python
#find the average of 10th grade reading by school
tenth_df = school_df.loc[school_df["grade"] == "10th", :]
tenth_df = tenth_df.rename(columns = {"reading_score": "10th"})
tenth_df = tenth_df.groupby("school_name")
tenth_read = tenth_df["10th"].mean()
tenth_read
```

```python
#find the average of 11th grade reading by school
eleventh_df = school_df.loc[school_df["grade"] == "11th", :]
eleventh_df = eleventh_df.rename(columns = {"reading_score": "11th"})
eleventh_df = eleventh_df.groupby("school_name")
eleventh_read = eleventh_df["11th"].mean()
eleventh_read
```

```python
#find the average of 12th grade reading by school
twelvth_df = school_df.loc[school_df["grade"] == "12th", :]
twelvth_df = twelvth_df.rename(columns = {"reading_score": "12th"})
twelvth_df = twelvth_df.groupby("school_name")
twelvth_read = twelvth_df["12th"].mean()
twelvth_read
```

```python
#put reading averages into a dataframe
Grade_df = pd.DataFrame([ninth_read, tenth_read, eleventh_read, twelvth_read])
Grade_df = Grade_df.T
Grade_df
```
